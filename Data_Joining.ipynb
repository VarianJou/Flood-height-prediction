# Input data

```python
# Define your input folder and list of CSVs
folder_path = 'data'
file_names = [
    'dem_features.csv',
    # 'HWM_Depth_m.csv',
    'combined_nlcd_features.csv',
    'precipitation.csv',
    'sentinel1_combined_features.csv',
    'USGS_HWM_Height.csv',
    'soil_moisture.csv',
    # 'Coordinates.csv'
    'weather_data.csv'
]
```

```python
# Check each file for presence of 'ID' column
for file in file_names:
    path = os.path.join(folder_path, file)
    df = pd.read_csv(path, nrows=1)  # Load just the header
    if 'ID' not in df.columns:
        print(f"❌ 'ID' column NOT found in: {file}")
    else:
        print(f"✅ 'ID' column found in: {file}")
```

```python


# Read and merge all files on 'ID' with custom suffixes
merged_df = pd.read_csv(os.path.join(folder_path, file_names[0]))

for i, file in enumerate(file_names[1:], 1):
    df = pd.read_csv(os.path.join(folder_path, file))
    merged_df = pd.merge(merged_df, df, on='ID', suffixes=('', f'_file{i}'))

merged_df
```

```python
combined_features = merged_df.drop(columns={'year','projection', 'nlcd_year' })



combined_features
```

```python
import numpy as np
from datetime import datetime, date
# Convert to datetime format
combined_features['peak_date'] = pd.to_datetime(combined_features['peak_date'])

combined_features['year'] = combined_features['peak_date'].dt.year
combined_features['month'] = combined_features['peak_date'].dt.month
combined_features['day'] = combined_features['peak_date'].dt.day
combined_features['hour'] = combined_features['peak_date'].dt.hour
combined_features


combined_features['month_sin'] = np.sin(2* np.pi * combined_features['month']/12)
combined_features['day_sin'] = np.sin(2* np.pi * combined_features['day']/31)
combined_features['hour_sin'] = np.sin(2* np.pi * combined_features['hour']/24)

combined_features=combined_features.drop(columns={'month', 'day', 'hour','peak_date', 'S1_Date', 'peak_date_file2', 'peak_date_file3', 'S1_Date_file5'})

combined_features = combined_features.select_dtypes(exclude=['datetime64'])
```

```python
from sklearn.impute import KNNImputer
import numpy as np
imputer = KNNImputer(n_neighbors=3)
df_imputed = imputer.fit_transform(combined_features)
```
